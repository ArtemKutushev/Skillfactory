# 8. Искусство визуализации

Познакомившись с потрясающим миром визуализации, вы, должно быть, поняли, что код построения графиков довольно простой и не требует специализированных навыков, документация ко всем библиотекам максимально пнятная и наполнена примерами. Построение графиков главныи образом заключается в настройке параметров их отображения.

    На самом деле все не так просто. Помимо изучения разичных типов графиков, нужно знать, для чего эти графики применяются, на каких данных их правильнее строить и как не потерять информативность, используя их.



### ОБЩИЕ РЕКОМЕНДАЦИИ К СОЗДАНИЮ ВИЗУАЛИЗАЦИИ

- 1) Первое и самое главное - ГРАФИК ДОЛЖЕН БЫТЬ ИНФОРМАТИВНЫМ И ПОНЯТНЫМ. То есть любой человек взглянув на график, должен понять что на нем изображено. Для этого всегда подписывайте оси графика или сам график, делайте интервалы между отметками на осях, не используйте слишком много графиков на одной координатной плоскости.

- 2) ОДНА ПЛОСКОСТЬ - ОДИН ВИД ГРАФИКОВ. Не стоит смешивать тип визуализации, это делает результат нечитабельным.

- 3) ПРИНЦИП МИНИМАЛИЗМА: ЧЕМ ПРОЩЕ ГРАФИК, ТЕМ ЛУЧШЕ - не нужно добавлять сглаживающие кривые, многочисленные подписи, лишние отмеки на осях, яркие, отвлекающие внимание цвета, если это не помогает вам донести идею.

- 4) НЕ ИСПОЛЬЗУЙТЕ СЛОЖНЫЙ ДИЗАЙН ТАМ, ГДЕ ЭТО НЕ ТРЕБУЕТСЯ. Прежде чем строить для презентации 3D-график, подумайте, можно ли обойтись без него.

- 5) ЕСЛИ ТИП ЗНАЧЕНИЙ ВСЕГО ОДИН, ЛЕГЕНДА НЕ НУЖНА. 

- 6) ЕСЛИ У ВАС НЕСКОЛЬК ДИАГРАММ, ИСПОЛЬЗУЙТЕ ЕДИНУЮ ЦВЕТОВУЮ ГАММУ.

- 7) ВРЕМЯ ВСЕГДА УКАЗЫВАЕТСЯ ПО ГОРИЗОНТАЛЬНОЙ ОСИ СЛЕВА НАПРАВО. Размещайте отмеки времени так, чтобы они не сливались друг с другом. *например, под углом 45 градусов).

- 8) ПРИ ПОСТРОЕНИИ ГРАФИКОВ НЕОБХОДИМО ОТТАЛКИВАТЬСЯ ОТ ИЗ ПРЕДНАЗНАЧЕНИЯ: например, не надо строить линейный график на категориальных данных - он не предназначен для этого.


### РЕКОМЕНДАЦИИ ПО ИСОЛЬЗОВАНЮ ПОПУЛЯРНЫХ ТИПОВ ДИАГРАММ

Навые визуализации предполагает правилный выбор диаграммы и умение ее преподнести. Давайте на примераз рассмотрим, как правильно (и неправльино тоже) строить основные виды графиков.


 #### СТОЛБЧАТАЯ ДИАГРАММА

Столбчатая диаграмма - один из самых популярных видов графиков, однако не все исследователи используют его правильно.

    Столбчатая диаграмма используется, когда необходимо сравнить какой-то показатель *количество, среднее, медиану) в зависимости от категориального признака (возможен -ва-риант сравнения по датам). Это часто помогает понять, -каково соотношение категорий, как категория доминирующая. Все это можно учитыать при построении модели, чтобы делать прогнозы более точными. 

##### СОВЕТЫ
    - Не сравнивать больше десяти категорий друг с другом с помощью стобчатой диаграммы - вам будет сложно ориентировать в ней, особенно если диаграмма ногоуровневая. Ели категорий слишком много, можно использовать иерарическую диаграмм или попробовать отразить диаграмму горизонтально (поменять парамтеры х и у местами). Еще один вагриант - объедининсть непопулярные категории в общую категорию "другие".

    - Начало точки отсчета для значений покателя всегда 0. Если у вас есть значения меньше 0, используйте горизонтальную столбчатую диаграмму. 

#### ЛИНЕЙНЫЙ ГРАФИК

    Линейный график отлично подходит, если набор данных непрерывен (как мы уже видели раньше, обычно это временной ряд). График используется для определения тенденций во временном ряду и сравнения нескольких рядов между собой.

##### СОВЕТЫ

    - Не используйте график, если набор анных дискретный (менее 20 наблюдений) - в таком случае лучше воспользуйтесь столбчатой диаграммой.

    - Время всегда отображается по оси асцисс и разбивается на равные интервалы.

    - Если даты сливаются, используйте наклон в 45 градусов.

    - Не используйте график для сравнения рядов, если их больше 7-10 - график станет нечитабельным. Попробуйте уменьшить число категорий. 

#### ГИСТОГРАММА

    Гистограммы часто применяюся для разведывательного анализа данных (EDA), тк они дают информацию о распредлении признака. С их помощью сразу определить диапазон изменения признака, его модальное значение (пик гистограммы), а также найти "пеньки", которые выбиваются от непрерывного распредлениея гистограммы, - аномалии.

##### СОВЕТЫ:

    - Не стоит строить гистограмму, есои наблюдений мало - распредлеение окажется далеким от действительного и вы просто сделаете ложные выводы. По статистике, для того, чтобы гистограмма хоть как-то оценивала истинное распределение, нужно как минимум 30 наблюдений (на практике нужно хтя бы 100).

    - Попробуйте построить пть-семь гистограмм на одном графике для и сравнения по категориям. На практике так делать не нужно. Для сравнения параметров распределений по категорям презназначена коробчатая диаграмма.

    - Если вы все же хотите сравнить гистограммы между собой, предварительно обязательно приведите призаки к одной шкале (мы делали это, когда сравнивали ежедневную заболеваемость коронавирусом в процентах от населени страны). Если этого не сделать, распределения окажутс несопоставимыми. После того как вы привели все признаки к одной шкале, лучше используйте цветовую гистограмму, которая показывает частоту интенсивностью цвета. Да, она не покажет высоту столбцов в цифрах, но зато вы сможете вищуально сравнить категории между собой и избежать наложени гистограмм друг на друга.

#### ДиИАГРАММА РАССЕЯНИЯ

    Диаграмма рассеяния и ее производные - предназначены для выявления взаимосвязи между признаками.

Добавлять в график расцветку по одному категориальнму признаку, а тип или размер маркеров - по другому. Итого мы сможем наблюдать взаимосвязь нескольких признаков (до пяти).

##### СОВЕТЫ:

    - Не используйте диаграммы рассния на маленьком наборе данных. Здесь ситуация та же, что и с гистограммами.

    - Не стоит использовать расцветку и размер точек для признаков с большим числом уникальных категорий. 

    - В случае если вы не видите зависимостей в данных, попробуйте использовать логарифическую шкалу по оси абсцисс (по оси абсцисс и ординат в случае 3D-графика). Во всех библиотеках в методе есть параметр log, значение которого нужно установить на TRUE. 

#### КРУГОВАЯ ДИАГРАММА

    Круговая диаграмма показывает структуру признака, то есть процентную долю каждого из возможных значений признака.

Вы сможете столкнуться с ней. например, когда захотите отобразить соотношение классов  данных или вклад отдельных компонентов в общую прибылть. 

##### СОВЕТЫ:

    - Суммы значений в круге всегда должна равняться единице, то есть всегда должно быть целое и его части.

    - Не визуализируйте секторы, близкие к 0, их невозможно сравнить друг с другом. 

    - Не делайте больше 6-8 секторов - воспринимать информацию будет сложно. Если компонентов болше, выделите топ-6-8, а остальные обозначьте как "прочие".

    - Всегда отображайте легенду либо подписи категорий внутри секторов.

    - Если важно выделить часть графика, "вытащите" его и центра.
